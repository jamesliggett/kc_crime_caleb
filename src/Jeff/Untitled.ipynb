```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Data Sets
df_14 = pd.read_csv('./Data/KCPD_Crime_Data_2014.csv')
df_15 = pd.read_csv('./Data/KCPD_Crime_Data_2015.csv')
df_16 = pd.read_csv('./Data/KCPD_Crime_Data_2016.csv')
df_17 = pd.read_csv('./Data/KCPD_Crime_Data_2017.csv')
df_18 = pd.read_csv('./Data/KCPD_Crime_Data_2018.csv')
```

```python
with plt.style.context('fivethirtyeight'):
    daily = df_vict_gun.set_index('Reported_Date').resample('W')['Firearm Used Flag'].count().plot(kind = 'line', 
                                                                                          figsize = [10, 6])
    daily.set_xlabel('')
    daily.set_ylabel('Number of Victims')
    daily.set_title('Number of Victims per Week Where Gun Involved')
    daily.text(x = 2270, y = -15,
    s = '   kc_crime_caleb                                                                                          Source: data.kcmo.org   ',
    fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey')
    plt.show()
```

```python
unique_reports = df.groupby('Reported_Date')['Report_No'].nunique()
```

```python
df1 = pd.DataFrame(unique_reports)
```

```python
df1.head()
```

```python
df1.reset_index(inplace = True)
```

```python
df1['Reported_Date'] = pd.to_datetime(df1['Reported_Date'])
```

```python
with plt.style.context('fivethirtyeight'):
    df1.set_index('Reported_Date').resample('D')['Report_No'].mean().plot(kind = 'line', figsize = [10,6],
                                                                         label = 'Daily')
    df1.set_index('Reported_Date').resample('W')['Report_No'].mean().plot(kind = 'line', 
                                                                         label = 'Weekly Average')
    month_avg = df1.set_index('Reported_Date').resample('M')['Report_No'].mean().plot(kind = 'line',
                                                                         label = 'Monthly Average')
    
    plt.legend()
    month_avg.text(x = 15900, y = 20,
    s = '   kc_crime_caleb                                                                                          Source: data.kcmo.org   ',
    fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey')
    month_avg.set_title('Crime for KCMO')
    month_avg.set_ylabel('Daily Crime Incidents')
    month_avg.set_xlabel('')
    plt.show()
```
